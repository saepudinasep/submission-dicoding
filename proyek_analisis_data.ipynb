## Pertanyaan Pertama : Bagaimana performa penjualan dan revenue perusahaan dalam beberapa bulan terakhir?

```python
# Merge tb_orders dengan tb_order_items berdasarkan kolom order_id
merged_orders_items = pd.merge(tb_orders, tb_order_items, on='order_id')

# Menggabungkan tb_orders, tb_order_items, dan tb_order_payments
merged_orders_items_payments = pd.merge(merged_orders_items, tb_order_payments, on='order_id')

monthly_orders_df = merged_orders_items_payments.resample(rule='M', on='order_purchase_timestamp').agg({
    "order_item_id": "nunique",
    "price": "sum"
})
monthly_orders_df.index = monthly_orders_df.index.strftime('%Y-%m')
monthly_orders_df = monthly_orders_df.reset_index()
monthly_orders_df.rename(columns={
    "order_item_id": "order_count",
    "price": "revenue"
}, inplace=True)
monthly_orders_df.head()
```

```python

monthly_orders_df = merged_orders_items_payments.resample(rule='M', on='order_purchase_timestamp').agg({
    "order_item_id": "nunique",
    "price": "sum"
})
monthly_orders_df.index = monthly_orders_df.index.strftime('%B')  # Mengubah format order date menjadi nama bulan

monthly_orders_df = monthly_orders_df.reset_index()
monthly_orders_df.rename(columns={
    "order_item_id": "order_count",
    "price": "revenue"
}, inplace=True)

plt.figure(figsize=(10, 5))
plt.plot(monthly_orders_df["order_purchase_timestamp"], monthly_orders_df["revenue"], marker='o', linewidth=2, color="#72BCD4")
plt.title("Number of Orders per Month", loc="center", fontsize=20)
plt.xlabel("Month ", fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```

## RFM Analysis dengan pertanyaan :
1. Kapan terakhir kali pelanggan dari kota tertentu melakukan transaksi?
2. Seberapa sering pelanggan dari kota tertentu melakukan pembelian dalam beberapa bulan terakhir?
3. Berapa total uang yang dihabiskan pelanggan dari kota tertentu dalam beberapa bulan terakhir?

```python
# Menggabungkan tb_orders dengan tb_customers
merged_orders_customers = pd.merge(tb_orders, tb_customers, on='customer_id')

merged_rfm = pd.merge(merged_orders_items_payments, merged_orders_customers, on='order_id')

# Mengubah kolom order_purchase_timestamp menjadi tipe datetime
merged_orders_customers['order_purchase_timestamp'] = pd.to_datetime(merged_orders_customers['order_purchase_timestamp'])

# Mencari tanggal transaksi terakhir pelanggan
last_purchase_date = merged_orders_customers.groupby('customer_unique_id')['order_purchase_timestamp'].max()

merged_rfm.info()
```

```python

# Terakhir
rfm_df = merged_rfm.groupby(by="customer_city", as_index=False).agg({
    "order_purchase_timestamp_x": "max", # mengambil tanggal order terakhir
    "order_id": "nunique", # menghitung jumlah order
    "payment_value": "sum" # menghitung jumlah revenue yang dihasilkan
})
rfm_df.columns = ["customer_city", "max_order_timestamp", "frequency", "monetary"]
# rfm_df.columns
# menghitung kapan terakhir pelanggan melakukan transaksi (hari)
rfm_df["max_order_timestamp"] = rfm_df["max_order_timestamp"].dt.date
recent_date = merged_rfm["order_purchase_timestamp_x"].dt.date.max()
rfm_df["recency"] = rfm_df["max_order_timestamp"].apply(lambda x: (recent_date - x).days)

rfm_df.drop("max_order_timestamp", axis=1, inplace=True)
rfm_df.head()
```

```python
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(40, 6))

colors = ["#72BCD4", "#72BCD4", "#72BCD4", "#72BCD4", "#72BCD4"]

sns.barplot(y="recency", x="customer_city", data=rfm_df.sort_values(by="recency", ascending=True).head(5), palette=colors, ax=ax[0])
ax[0].set_ylabel(None)
ax[0].set_xlabel(None)
ax[0].set_title("By Recency (days)", loc="center", fontsize=18)
ax[0].tick_params(axis ='x', labelsize=15)

sns.barplot(y="frequency", x="customer_city", data=rfm_df.sort_values(by="frequency", ascending=False).head(5), palette=colors, ax=ax[1])
ax[1].set_ylabel(None)
ax[1].set_xlabel(None)
ax[1].set_title("By Frequency", loc="center", fontsize=18)
ax[1].tick_params(axis='x', labelsize=15)

sns.barplot(y="monetary", x="customer_city", data=rfm_df.sort_values(by="monetary", ascending=False).head(5), palette=colors, ax=ax[2])
ax[2].set_ylabel(None)
ax[2].set_xlabel(None)
ax[2].set_title("By Monetary", loc="center", fontsize=18)
ax[2].tick_params(axis='x', labelsize=15)

plt.suptitle("Best City Based on RFM Parameters (customer_city)", fontsize=20)
plt.show()
```
